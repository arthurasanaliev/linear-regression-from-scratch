# Part 1: Linear Regression with One Variable

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def get_input():
    n = int(input("Enter number of points: "))
    points = []
    
    print("x y")
    for _ in range(n):
        x, y = map(int, input().split())
        points.append([x, y])
        
    return points
```

```python
def gradient_descent(m_now, b_now, points):
    L = 0.001
    m_der, b_der = 0, 0
    n = len(points)
    
    for point in points:
        x, y = point[0], point[1]
        m_der += -(2/n) * x * (y - (m_now * x + b_now))
        b_der += -(2/n) * (y - (m_now * x + b_now))
        
    m = m_now - L * m_der
    b = b_now - L * b_der
    
    return m, b
```

```python
def loss(m, b, points):
    total_loss = 0
    n = len(points)
    
    for point in points:
        x, y = point[0], point[1]
        total_loss += (y - (m * x + b)) ** 2
    
    return total_loss / n
```

```python
def plot(m, b, points):
    xs, ys = [], []
    for point in points:
        xs.append(point[0])
        ys.append(point[1])
    
    plt.scatter(xs, ys, color='black')
    x_range = range(min(xs), max(xs) + 1)
    plt.plot(x_range, [m * x + b for x in x_range], color='red', label='Regression Line')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.legend()
    plt.show()
```

```python
def plot_loss_function(m, b, points):
    m_range = np.linspace(-10, 10, 1000)
    b_range = np.linspace(-100, 100, 1000)
    M, B = np.meshgrid(m_range, b_range)
    Z = np.array([[loss(m, b, points) for m, b in zip(row_m, row_b)] for row_m, row_b in zip(M, B)])
    
    plt.figure(figsize=(10, 6))
    plt.contourf(M, B, Z, levels=20, cmap='viridis')
    plt.colorbar(label='Loss')
    plt.xlabel('m')
    plt.ylabel('b')
    plt.title('Loss Function')
    plt.scatter(m, b, color='red')
    plt.pause(0.1)
```

```python
def start(points):
    m, b = 0, 0
    epochs = 10001
    
    for i in range(epochs):
        m, b = gradient_descent(m, b, points)
#         if i % 2000 == 0:
#             print(f"{i} gives:")
#             print(m, b)
#             plot_loss_function(m, b, points)

    print(f"Regression Line: y = {m}x + {b}")
    plot(m, b, points)
```

```python
def main():
    points = get_input()
    start(points)
```

```python
main() 
```
